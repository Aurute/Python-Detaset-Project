```python
def Amziaus_rezis(eilute):
    if -1 <= eilute['age'] and eilute['age'] <= 10:
        return "Labai jauna kompanija"
    elif 11 <= eilute['age'] and eilute['age'] <= 50:
        return "Jauna kompanija"
    elif 51 <= eilute['age'] and eilute['age'] <= 100:
        return "Vidutinė kompanija"
    elif 101 <= eilute['age'] and eilute['age'] <= 200:
        return "Labai sena kompanija"
    elif 201 <= eilute['age'] and eilute['age'] <= 300:
        return "Itin sena kompanija"
    else:
        return "Amžius už galimų ribų"
```

```python
top_companies_size_age = JB.groupby('age').size().sort_values(ascending=False).head(10)
print(top_companies_size_age)
```

```python
colors = ['skyblue', 'salmon', 'lightgreen']

top_companies_size_age.plot(kind='bar', color=colors, figsize=(10, 4))

plt.xlabel('Company')
plt.ylabel('Size / Age')
plt.title('Company Size and Age')

plt.tight_layout()
plt.show()
```

```python
JB['age'] = JB.apply(Amziaus_rezis, axis=1)
JB.head(10)
```

```python
job_titles = ['Data Scientist', 'Data Engineer', 'Data Analyst', 'Senior Data Analyst', 'Senior Data Engineer', 'Senior Data Scientist']
counts = [131, 53, 15, 20, 25, 30]

colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'purple', 'lightcoral']

plt.figure(figsize=(8, 4))
bars = plt.bar(job_titles, counts, color=colors)

plt.xlabel('Job Title')
plt.ylabel('Count')
plt.title('Number of Each Job Title')
plt.xticks(rotation=45, ha='right')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')  # Pridedame skaičių virš stulpelio

plt.tight_layout()
plt.show()
```

```python
top_specialybes = JB['Job Title'].value_counts().sort_values(ascending=False).head(10)
top_specialybes
```

```python
ax = top_specialybes.plot(kind='bar', color=plt.cm.Paired(range(len(top_specialybes))))
for i, v in enumerate(top_specialybes):
    ax.text(i, v + 0.1, str(v), ha='center', va='bottom')

ax.set_title('Dažniausiai pasikartojančios specialybės')
ax.set_xlabel('Specialybė')
ax.set_ylabel('Dažnis')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(top_specialybes, labels=top_specialybes.index, autopct='%1.1f%%', colors=plt.cm.Paired(range(len(top_specialybes))))

ax.set_title('Dažniausiai pasikartojančios specialybės')

plt.tight_layout()
plt.show()
```
